A biblioteca `matplotlib` é uma poderosa ferramenta para criação de gráficos em Python. Ela oferece uma variedade de funções e classes para criar diferentes tipos de gráficos e personalizá-los de acordo com suas necessidades. Aqui estão algumas das principais funções e classes da `matplotlib`:

1. **`pyplot`:** O módulo `pyplot` é a interface mais comum para criar gráficos em `matplotlib`. Ele fornece funções semelhantes às encontradas no MATLAB para criar e personalizar gráficos.

   Exemplo:
   ```python
   import matplotlib.pyplot as plt
   plt.plot(x, y, label='Dados de Exemplo')
   plt.xlabel('Eixo X')
   plt.ylabel('Eixo Y')
   plt.legend()
   plt.show()
   ```

2. **`figure`:** A classe `figure` é usada para criar uma figura (janela) na qual você pode adicionar gráficos. É útil quando você deseja criar vários gráficos em uma única figura.

   Exemplo:
   ```python
   fig = plt.figure()
   ax1 = fig.add_subplot(2, 2, 1)  # Cria um subplot na posição (2, 2, 1)
   ```

3. **`subplot`:** A função `subplot` permite criar uma grade de subplots em uma única figura. Ela é frequentemente usada em conjunto com `figure`.

   Exemplo:
   ```python
   plt.subplot(2, 2, 1)  # Cria um subplot na posição (2, 2, 1)
   ```

4. **`bar` e `barh`:** Essas funções são usadas para criar gráficos de barras verticais e horizontais, respectivamente.

   Exemplo:
   ```python
   plt.bar(x, y, label='Dados de Exemplo')
   ```

5. **`hist`:** A função `hist` é usada para criar um histograma a partir de dados.

   Exemplo:
   ```python
   plt.hist(data, bins=10, alpha=0.5, label='Histograma')
   ```

6. **`scatter`:** A função `scatter` é usada para criar gráficos de dispersão, onde você pode plotar pontos individuais.

   Exemplo:
   ```python
   plt.scatter(x, y, label='Pontos de Exemplo', color='red', marker='o')
   ```

7. **`plot`:** A função `plot` é usada para criar gráficos de linhas, onde você pode traçar uma série de pontos e conectá-los com linhas.

   Exemplo:
   ```python
   plt.plot(x, y, label='Dados de Exemplo', linestyle='--', marker='o', color='green')
   ```

8. **`imshow`:** A função `imshow` é usada para exibir imagens.

   Exemplo:
   ```python
   plt.imshow(imagem)
   ```

9. **`legend`:** A função `legend` é usada para adicionar legendas aos gráficos.

   Exemplo:
   ```python
   plt.legend()
   ```

10. **`xlabel` e `ylabel`:** Essas funções são usadas para adicionar rótulos aos eixos X e Y, respectivamente.

    Exemplo:
    ```python
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo Y')
    ```

11. **`title`:** A função `title` é usada para adicionar um título ao gráfico.

    Exemplo:
    ```python
    plt.title('Título do Gráfico')
    ```

12. **`grid`:** A função `grid` é usada para adicionar grades ao gráfico.

    Exemplo:
    ```python
    plt.grid(True)
    ```

Essas são apenas algumas das principais funções e classes da biblioteca `matplotlib`. Você pode explorar a documentação oficial para aprender mais sobre suas capacidades e como usá-las para criar gráficos personalizados: https://matplotlib.org/stable/contents.html

````python
from matplotlib import pyplot as plt
import numpy as np
import math
x=np.arange(0, math.pi*2, 0.05)
y=np.sin(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()

````

````python
import matplotlib.pyplot as plt
import numpy as np

x=['python','JS','JAVA', 'C', 'C++','C#']#definir a amostra que ficará na horizontal
y=[33,88,122,55,60,22]#altura de cada bagulho na vertical
z=[12,33,2,1,4,66,5]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('linguagens', fontsize=15)
plt.ylabel('pontuação',fontsize=15)
plt.title('RANKING',fontsize=15)

plt.bar(p,y,width,color='r',label='pop')
plt.bar(p1,z,width,color='y',label='pop1')

plt.legend()
plt.show()
````

````python
from scipy.integrate import quad
import numpy as np

def normal_fdp(x, mu, sigma_2):
    return np.exp(-(x - mu) ** 2 / (2 * sigma_2)) / np.sqrt(2 * np.pi * sigma_2)

standard_fdp = lambda x: normal_fdp(x, 0, 1)
integral = quad(standard_fdp, -3, 3)
print('P(-3 < X < 3): {}, erro: {}'.format(integral[0], integral[1]))
````
